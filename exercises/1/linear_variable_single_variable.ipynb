```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.tail()
```

```python
plt.xlabel('year', fontsize = 20)
plt.ylabel('per capita income (US$)', fontsize = 20)
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
X = df[['year']]
y = df['per capita income (US$)']
y
```

```python
model.fit(X,y)
```

```python
model.predict([[2020]])
```

```python
plt.xlabel('year', fontsize = 20)
plt.ylabel('per capita income (US$)', fontsize = 20)
plt.scatter(df['year'], df['per capita income (US$)'], color='red', marker='+')
plt.plot(df['year'], model.predict(df[['year']]), color='blue')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df[['year']], df['per capita income (US$)'], train_size = 0.8, random_state=10)
```
