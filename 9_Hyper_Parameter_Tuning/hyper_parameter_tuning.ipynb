```python

import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df[20:37]
```

```python
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df[47:54]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf', C=30, gamma='auto')
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='linear', C=10, gamma='auto'), iris.data, iris.target, cv = 5)
```

```python
cross_val_score(SVC(kernel='rbf', C=10, gamma='auto'),
                iris.data, iris.target, cv=5)
```

```python
cross_val_score(SVC(kernel='linear', C=20, gamma='auto'),
                iris.data, iris.target, cv=5)
```

```python
import numpy as np
kernals = ['rbf','linear']
C = [1,10,20]
avg_score = {}

for kval in kernals:
    for cval in C:
        cv_score = cross_val_score(
            SVC(kernel=kval, C=cval, gamma='auto'), iris.data, iris.target, cv=5)
        avg_score[kval +'_'+ str(cval)] = np.average(cv_score)

avg_score
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma='auto'), {
    'C':[1,10,20],
    'kernel':['rbf','linear']
}, cv=5, return_train_score=False)

clf.fit(iris.data, iris.target)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(SVC(gamma='auto'),
                        {
                            'C':[1,10,20],
                            'kernel':['rbf','linear']
                        }, cv=5, n_iter=2, return_train_score=False)
rs.fit(X_train,y_train)
pd.DataFrame(rs.cv_results_)[['param_C', 'param_kernel', 'mean_test_score']]
```

Now choosing the best model for the problem

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm
```

```python
# A dictionary of models
model_params = {
    'svm': {
        'model' : svm.SVC(gamma='auto'),
        'params' : {
            'C' : [1,10,20],
            'kernel' : ['rbf', 'linear']
        }
    },
    'random_forest' : {
        'model' : RandomForestClassifier(),
        'params' : {
            'n_estimators' : [1,5,10]
        }
    },
    'logistic_regression' : {
        'model' : LogisticRegression(),
        'params' : {
            'C' : [1,5,10]
        }
    }
}
```

```python
model_params.items()
```

```python
scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(iris.data, iris.target)
    scores.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_params' : clf.best_params_
    })
```

```python
pd.DataFrame(scores)
```
